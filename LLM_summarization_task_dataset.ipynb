```python
import pandas as pd


df = pd.read_csv("/home/bs_thesis/Desktop/Rushi/BS_thesis/survey dataset of task dataset.csv")
```

```python
cols = ["Read_skill_gap", "Write_skill_gap", "Speak_skill_gap", 
        "Lect_skill_gap", "conv_skill_gap"]

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
cluster_summary = df.groupby("cluster")[cols].mean()
print(cluster_summary)
```

```python
def level(x):
    if x > 2:
        return "high"
    elif x > 1:
        return "moderate"
    else:
        return "low"

text_data = ""

for cluster, row in cluster_summary.iterrows():
    text_data += f"""
    Cluster {cluster} shows {level(row.mean())} overall skill gap.
    Reading gap is {level(row['Read_skill_gap'])},
    writing gap is {level(row['Write_skill_gap'])},
    speaking gap is {level(row['Speak_skill_gap'])},
    lecture gap is {level(row['Lect_skill_gap'])},
    and conversation gap is {level(row['conv_skill_gap'])}.
    """
```

```python
prompt = f"""
You are an expert data analyst.

Below is cluster-wise student skill gap analysis:

{text_data}

Analyze and summarize the data clearly.

Answer the following:
1. Which skill is the most difficult?
2. Which skill is the least difficult?
3. Rank the skills from most to least difficult.
4. What overall learning pattern do you observe?
5. What major challenges do students face?
6. What factors contribute to these difficulties?
7. Provide 2–3 key insights.
8. What improvements would you suggest for students?

Write the answer in simple, clear sentences.
"""
```
