```python
import pandas as pd

df = pd.read_csv("/home/bs_thesis/Desktop/Rushi/BS_thesis/survey dataset of task dataset.csv")

df.head()
```

```python
cols = ["Read_skill_gap", "Write_skill_gap", "Speak_skill_gap", 
        "Lect_skill_gap", "conv_skill_gap"]

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
cluster_summary = df.groupby("cluster")[cols].mean()
print(cluster_summary)
```

```python
text_data = ""

for cluster, row in cluster_summary.iterrows():
    text_data += f"""
    Cluster {cluster} shows:
    Reading: {row['Read_skill_gap']:.2f}
    Writing: {row['Write_skill_gap']:.2f}
    Speaking: {row['Speak_skill_gap']:.2f}
    Lecture: {row['Lect_skill_gap']:.2f}
    """
```

```python
regression_insight = """
The main factors affecting skill difficulty are:
- Mother Tongue Influence (strong positive effect)
- Difficulty in understanding movies even with subtitle
- Difficulty in switching to English
"""

final_text = text_data + regression_insight
```

```python
prompt = f"""
Analyze the following cluster-wise student difficulty data:

{final_text}

Answer:
1. Which cluster has highest difficulty?
2. What skills are most difficult?
3. Why are students facing these difficulties?
4. What patterns do you observe?
5. Which skill is the least difficult?
6. Rank the skills from most to least difficult.
7. What overall learning pattern do you observe?
8. What major challenges do students face?
9 What factors contribute to these difficulties?
10. Provide 2–3 key insights.
11. What improvements would you suggest for students?
"""
```

```python
from groq import Groq
client = Groq(api_key="API_KEY_HERE")

chat = client.chat.completions.create(
    model="llama-3.1-8b-instant", 
    messages=[{"role": "user", "content": prompt}]
)
```
